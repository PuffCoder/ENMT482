```python
# Michael P. Hayes UCECE, Copyright 2018--2019
import numpy as np
import matplotlib.pyplot as plt  # Add import for matplotlib
from ipywidgets import interact, interactive, fixed

# Import style module from matplotlib
import matplotlib.style as style
style.use('default')  # Graph white background 

# Define the signal_plot function or import it from 'lib.signal_plot'

def linearise_demo1_plot(x0=(0,10)):

    Nx = 201
    x = np.linspace(0, 10, Nx)

    a0 = 10
    a1 = 1
    
    h = a0 / (a1 + x)
    # h = a0 / (a1 + x) + b * x + c

    # H 方程的倒数/斜率
    J = -a0 / (a1 + x0)**2 

    h2 = J * (x - x0) + a0 / (a1 + x0) 
    # + b * x0 + c

    fig = plt.figure()  # Create a new figure
    ax = fig.add_subplot(111)  # Add a subplot
    ax.plot(x, h)  # Plot the original function
    ax.plot(x, h2)  # Plot the linearized function
    ax.set_xlabel('$x$')    
    plt.show()  # Show the plot
    
linearise_demo1_plot(x0=1)  # Pass x0 as an argument
```

```python
# Michael P. Hayes UCECE, Copyright 2018--2019
import numpy as np
from ipywidgets import interact, interactive, fixed
from lib.signal_plot import signal_plot

def linearise_demo2_plot(x0=1):

    Nx = 201
    x = np.linspace(0, 10, Nx)

    a = 1
    b = -9
    c = 27
    d = 100

    h = a * x**3 + b * x**2 + c * x + d
    
    J = 3 * a * x0**2 + 2 * b * x0 + c
    u = a * x0**3 + b * x0**2 + c * x0 + d    
    h2 = (x - x0) * J + u

    fig = signal_plot(x, h)
    ax = fig.axes[0]
    ax.set_xlabel('$x$')
    ax.plot(x, h2)
    ax.set_ylim(0, max(max(h), max(h2)))
    

def linearise_demo2():
    interact(linearise_demo2_plot, x0=(0, 10))


linearise_demo2()
```
